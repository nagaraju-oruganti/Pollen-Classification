```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
path = '../models/aug_0.5_v2/results.pkl'
with open(path, 'rb') as f:
    results = pickle.load(f)
    
cols = ['epoch', 'train_loss', 'valid_loss', 'train_f1', 'valid_f1', 'train_acc', 'valid_acc', 'lr']
df = pd.DataFrame(results, columns = cols)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 4))
# loss plot
sns.lineplot(data = df, x = 'epoch', y = 'train_loss', label = 'Train loss', marker = 'o', ax = ax[0])
sns.lineplot(data = df, x = 'epoch', y = 'valid_loss', label = 'Validation loss', marker = 'o', ax  = ax[0])
# accuracy plot
sns.lineplot(data = df, x = 'epoch', y = 'train_acc', label = 'Train accuracy', marker = 'o', ax = ax[1])
sns.lineplot(data = df, x = 'epoch', y = 'valid_acc', label = 'Validation accuracy', marker = 'o', ax  = ax[1])

ax[0].set_ylabel('Loss')
ax[1].set_ylabel('Accuracy')

plt.show()
```
